```python
def cleaned_price(x):
    x = str(x).replace("Cr" , "").replace('₹' , "").strip()
    return float(x)
```

```python
df_price_cleaned = [round(cleaned_price(x), 2) for x in df_price]
```

```python
df_project = df_mod["Project_Name"].to_list()
```

```python
def clean_project_name(x):
    x = str(x).replace("Project_" , "").strip()
    return int(x)
```

```python
df_project_cleaned = [clean_project_name(x) for x in df_project]
```

```python
df_project_cleaned
```

```python
df_mod['Configuration'].to_list()
```

```python
def clean_configuration(x):
    if '+' in str(x):
        x = x.replace('+' , '')
    x = x.upper()
    return x
```

```python
df_cleaned_configuration = [clean_configuration(x) for x in df_raw['Configuration'].to_list()]
```

```python
df_cleaned_configuration
```

```python
df_mod['Configuration'] = df_cleaned_configuration
```

```python
# Fill null Amenity_Score with median of its Micro_Market
df_mod['Amenity_Score'] = df_mod.groupby('Micro_Market')['Amenity_Score'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df_mod['Booking_status'] = df_mod['Buyer_Comments'].apply(
    lambda x: 'Not Yet Booked' if pd.isna(x) else 'Booked'
)
```
